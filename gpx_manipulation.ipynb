```python
import pandas as pd
```

```python
from gpxcsv import gpxtolist
```

```python
gpx_list = gpxtolist('gpx_files/20250825.gpx')

#if you have pandas
df = pd.DataFrame(gpx_list)
```

```python
df = df.drop('type', axis = 1)
df = df.drop('name', axis = 1)
df = df.drop('ele', axis = 1)
```

```python
def clean_time(time):
    time = time.replace('-', '')
    time = time.replace('T', '')
    time = time.replace(':', '')
    time = time.replace('Z', '')
    return time
```

```python
df['time'] = df['time'].apply(clean_time)
```
